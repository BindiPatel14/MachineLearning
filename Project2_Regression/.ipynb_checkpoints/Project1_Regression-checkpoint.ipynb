<b>REGRESSION: Predict Laptop Prices</b>

The data for the regression part of my project has been taken from Kaggle. The URL for the data set is as follows:

URL : https://www.kaggle.com/muhammetvarl/laptop-price

Description of the dataset: This dataset consist of laptop price observations depending on various features.<br><br>
Outcome Variable: <br>
Price - Predict the price of a laptop

<b>In this project, I am running below machine learning models to predict the laptop prices.

KNN Regressor<br>
Linear Regression<br>
Ridge<br>
Lasso<br>
Polynomial Features<br>
Decision Tree Regressor <br>
Support Vector Machine Regressor with kernel (rbf, poly and linear)<br></b>

Features: 12<br>
Instances: 1303<br>
Missing data: No<br>

FEATURE DESCRIPTION:

Company- String -Laptop Manufacturer<br>
Product -String -Brand and Model<br>
TypeName -String -Type (Notebook, Ultrabook, Gaming, etc.)<br>
Inches -Numeric- Screen Size<br> 
ScreenResolution -String- Screen Resolution<br> 
Cpu- String -Central Processing Unit (CPU)<br> 
Ram -String- Laptop RAM<br> 
Memory -String- Hard Disk / SSD Memory<br> 
GPU -String- Graphics Processing Units (GPU)<br> 
OpSys -String- Operating System<br> 
Weight -String- Laptop Weight<br> 
Price_euros -Numeric- Price (Euro)

**Data Scaling**

```python
df['Price_euros']=np.log(df['Price_euros'])
```

```python
df['dimen'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)')
df['Type_of_screen'] = df['ScreenResolution'].replace(r'(\d+x\d+)','',regex=True)
df['Type_of_screen'] = df['Type_of_screen'].replace(r'(Full HD|Quad HD|Quad HD|\+|/|4K Ultra HD)','',regex=True)
df['touch_screen'] = df['Type_of_screen'].str.extract(r'(Touchscreen)')
df['Type_of_screen'] = df['Type_of_screen'].replace(r'(Touchscreen)','',regex=True)
df['touch_screen'] = df['touch_screen'].replace('Touchscreen',1)
df['touch_screen'] = df['touch_screen'].replace(np.nan,0)
df['Type_of_screen']=df['Type_of_screen'].replace(r' ','',regex=True)
df['Type_of_screen'] = df['Type_of_screen'].replace(r'^\s*$', np.nan, regex=True)
```

```python
df['freq_of_cpu'] = df['Cpu'].str.extract(r'(\d+(?:\.\d+)?GHz)') 
df['freq_of_cpu'] = df['freq_of_cpu'].str.replace('GHz', '')
df['Cpu']= df['Cpu'].str.replace(r'(\d+(?:\.\d+)?GHz)', '', regex=True)
```

```python
df['Ram'] = df['Ram'].str.replace('GB', '')
df['Ram'] = df['Ram'].astype(int)
```

```python
df['Weight'] = df['Weight'].str.replace('kg','').astype(float)
```

```python
df['Memory'] = df['Memory'].replace(['128GB SSD', '128GB Flash Storage', '128GB HDD'], '128')
df['Memory'] = df['Memory'].replace(['256GB SSD', '256GB Flash Storage'], '256')
df['Memory'] = df['Memory'].replace(['512GB SSD', '256GB SSD +  256GB SSD', '512GB Flash Storage'], '512')
df['Memory'] = df['Memory'].replace(['500GB HDD'], '500')
df['Memory'] = df['Memory'].replace(['1TB HDD', '1TB SSD', '1.0TB Hybrid', '512GB SSD +  512GB SSD', '1.0TB HDD'], '1024')
df['Memory'] = df['Memory'].replace(['2TB HDD', '1TB SSD +  1TB HDD', '1TB HDD +  1TB HDD',], '2048')
df['Memory'] = df['Memory'].replace(['128GB SSD +  2TB HDD'], '2160')
df['Memory'] = df['Memory'].replace(['32GB Flash Storage', '32GB SSD', '32GB HDD'], '32')
df['Memory'] = df['Memory'].replace(['256GB SSD +  500GB HDD'], '756')
df['Memory'] = df['Memory'].replace(['16GB Flash Storage', '16GB SSD'], '16')
df['Memory'] = df['Memory'].replace(['128GB SSD +  1TB HDD'], '1148')
df['Memory'] = df['Memory'].replace(['64GB Flash Storage', '64GB SSD'], '64')
df['Memory'] = df['Memory'].replace(['180GB SSD'], '180')
df['Memory'] = df['Memory'].replace(['240GB SSD'], '240')
df['Memory'] = df['Memory'].replace(['512GB SSD +  256GB SSD'], '768')
df['Memory'] = df['Memory'].replace(['256GB SSD +  1TB HDD', '256GB SSD +  1.0TB Hybrid'], '1280')
df['Memory'] = df['Memory'].replace(['512GB SSD +  1TB HDD', '512GB SSD +  1.0TB Hybrid'], '1536')
df['Memory'] = df['Memory'].replace(['256GB SSD +  2TB HDD'], '2304')
df['Memory'] = df['Memory'].replace(['128GB SSD +  2TB HDD'], '2176')
df['Memory'] = df['Memory'].replace(['512GB SSD +  2TB HDD'], '2560')
df['Memory'] = df['Memory'].replace(['8GB SSD'], '8')
df['Memory'] = df['Memory'].replace(['508GB Hybrid'], '508')
df['Memory'] = df['Memory'].replace(['64GB Flash Storage +  1TB HDD'], '1088')
```

```python
df['Gpu'] = df['Gpu'].str.extract(r'^(\w+)')
```

**Drop unnecessary columns**

```python
df=df.drop(['ScreenResolution','laptop_ID','Type_of_screen'],axis=1)
```

```python
en = LabelEncoder()
df['Company']= en.fit_transform(df['Company'])
df['Product']= en.fit_transform(df['Product'])
df['TypeName']= en.fit_transform(df['TypeName'])
df['Cpu']= en.fit_transform(df['Cpu'])
df['OpSys']= en.fit_transform(df['OpSys'])
df['Gpu']= en.fit_transform(df['Gpu'])
df['dimen']= en.fit_transform(df['dimen'])
```

**Justify the type of scaling**

```python
fig, ax = plt.subplots(figsize=(14,9))
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu", ax=ax)
```

Observe the relationship between all columns by using heatmaps.

Split dataset into two parts

```python
X=df.drop('Price_euros', axis = 1).values
```

```python
y=df['Price_euros'].values
```

```python
###Take 30 percent for testing and 70 percent for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 69)
```

StandardScaler

```python
#####Apply standard Scaler for normalizing all columns to the same scale
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Apply models with default hyper parameters and then do cross validation and then apply Grid search for selecting the best hyperparameters.

## KNN Regressor

```python
knn = neighbors.KNeighborsRegressor(n_neighbors = 5)
knn.fit(X_train, y_train)
pred_knn=knn.predict(X_test)
print(knn.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_knn))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_knn))
print("RMSE using KNNeighborsRegressor: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(knn, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
parameters = [{'weights': ['uniform'], 'n_neighbors': [1,2,3,5, 10, 20]}]
knn_gscv = GridSearchCV(knn, param_grid=parameters, cv=5)
knn_gscv.fit(X_train, y_train)
knn_gscv.best_params_
```

## Linear Regression

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
pred_regr=regr.predict(X_test)
print(regr.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_regr))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_regr))
print("RMSE using LinearRegression: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(regr, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
regr_gscv = GridSearchCV(regr, param_grid=parameters, cv=5)
regr_gscv.fit(X_train, y_train)
regr_gscv.best_params_
```

## Ridge

```python
ridge= Ridge()
ridge.fit(X_train, y_train)
pred_ridge=ridge.predict(X_test)
print(ridge.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_ridge))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_ridge))
print("RMSE using Ridge: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(ridge, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
parameters = dict(alpha=alpha)
ridge_gscv = GridSearchCV(ridge, param_grid=parameters, cv=5)
ridge_gscv.fit(X_train, y_train)
ridge_gscv.best_params_
```

## Lasso

```python
lasso= Lasso(alpha=0.02)
lasso.fit(X_train, y_train)
pred=lasso.predict(X_test)
print(lasso.score(X_train,y_train))
print(lasso.score(X_test,y_test))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred))
print("RMSE using Lasso: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(lasso, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
lasso_params = {'alpha':[0.02, 0.024, 0.025, 0.026, 0.03]}
lasso_gscv = GridSearchCV(lasso, param_grid=lasso_params, cv=5)
lasso_gscv.fit(X_train, y_train)
lasso_gscv.best_params_
```

```python
polyreg = make_pipeline(PolynomialFeatures(2) , linear_model.LinearRegression())
polyreg.fit(X_train,y_train)
pred_polyreg=polyreg.predict(X_test)
print(polyreg.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_polyreg))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_polyreg))
print("RMSE using PolynomialFeatures: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(polyreg, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
param_grid = {'polynomialfeatures__degree': np.arange(5), 'linearregression__fit_intercept': [True, False], 'linearregression__normalize': [True, False]}

polyreg_gscv= GridSearchCV(polyreg, param_grid, cv=5, scoring='neg_mean_squared_error')
polyreg_gscv.fit(X_train, y_train)
polyreg_gscv.best_params_
```

## Decision Tree Regressor

```python
Dec_reg=DecisionTreeRegressor()
Dec_reg.fit(X_train, y_train)
pred_dt=Dec_reg.predict(X_test)
print(Dec_reg.score(X_train,y_train))
print("Score of model: ",r2_score(y_test,pred_dt))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_dt))
print("RMSE using DecisionTreeRegressor: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
```

```python
scores = cross_val_score(Dec_reg, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
params = {'max_depth': [3,4,5,6], 'min_samples_leaf': [0.02,0.03,0.04,0.06,0.08,0.3,0.4]}
Dec_reg_gscv = GridSearchCV(Dec_reg, param_grid=params, cv=5)
Dec_reg_gscv.fit(X_train, y_train)
Dec_reg_gscv.best_params_
```
