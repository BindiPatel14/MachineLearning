**Data Scaling**

```python
# Scaling the price column into a range
df['Price_euros']=np.log(df['Price_euros'])

# Extract the only dimention values from the each column values 
df['dimen'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)')

# Extract the only dimention values from the Type_of_screen and 'touch_screen'
df['Type_of_screen'] = df['ScreenResolution'].replace(r'(\d+x\d+)','',regex=True)
df['Type_of_screen'] = df['Type_of_screen'].replace(r'(Full HD|Quad HD|Quad HD|\+|/|4K Ultra HD)','',regex=True)
df['touch_screen'] = df['Type_of_screen'].str.extract(r'(Touchscreen)')
df['Type_of_screen'] = df['Type_of_screen'].replace(r'(Touchscreen)','',regex=True)
df['touch_screen'] = df['touch_screen'].replace('Touchscreen',1)
df['touch_screen'] = df['touch_screen'].replace(np.nan,0)
df['Type_of_screen']=df['Type_of_screen'].replace(r' ','',regex=True)
df['Type_of_screen'] = df['Type_of_screen'].replace(r'^\s*$', np.nan, regex=True)

# Extract the only numeric values from the freq_of_cpu column
df['freq_of_cpu'] = df['Cpu'].str.extract(r'(\d+(?:\.\d+)?GHz)') 
df['freq_of_cpu'] = df['freq_of_cpu'].str.replace('GHz', '')
df['Cpu']= df['Cpu'].str.replace(r'(\d+(?:\.\d+)?GHz)', '', regex=True)

# Remove the GB Tag from the all values of the Ram column
df['Ram'] = df['Ram'].str.replace('GB', '')
df['Ram'] = df['Ram'].astype(int)

# Remove the Kg from all the values of the Weight column
df['Weight'] = df['Weight'].str.replace('kg','').astype(float)

# Remove all the textual lines from the Memory column and leave the number of the value only
df['Memory'] = df['Memory'].replace(['128GB SSD', '128GB Flash Storage', '128GB HDD'], '128')
df['Memory'] = df['Memory'].replace(['256GB SSD', '256GB Flash Storage'], '256')
df['Memory'] = df['Memory'].replace(['512GB SSD', '256GB SSD +  256GB SSD', '512GB Flash Storage'], '512')
df['Memory'] = df['Memory'].replace(['500GB HDD'], '500')
df['Memory'] = df['Memory'].replace(['1TB HDD', '1TB SSD', '1.0TB Hybrid', '512GB SSD +  512GB SSD', '1.0TB HDD'], '1024')
df['Memory'] = df['Memory'].replace(['2TB HDD', '1TB SSD +  1TB HDD', '1TB HDD +  1TB HDD',], '2048')
df['Memory'] = df['Memory'].replace(['128GB SSD +  2TB HDD'], '2160')
df['Memory'] = df['Memory'].replace(['32GB Flash Storage', '32GB SSD', '32GB HDD'], '32')
df['Memory'] = df['Memory'].replace(['256GB SSD +  500GB HDD'], '756')
df['Memory'] = df['Memory'].replace(['16GB Flash Storage', '16GB SSD'], '16')
df['Memory'] = df['Memory'].replace(['128GB SSD +  1TB HDD'], '1148')
df['Memory'] = df['Memory'].replace(['64GB Flash Storage', '64GB SSD'], '64')
df['Memory'] = df['Memory'].replace(['180GB SSD'], '180')
df['Memory'] = df['Memory'].replace(['240GB SSD'], '240')
df['Memory'] = df['Memory'].replace(['512GB SSD +  256GB SSD'], '768')
df['Memory'] = df['Memory'].replace(['256GB SSD +  1TB HDD', '256GB SSD +  1.0TB Hybrid'], '1280')
df['Memory'] = df['Memory'].replace(['512GB SSD +  1TB HDD', '512GB SSD +  1.0TB Hybrid'], '1536')
df['Memory'] = df['Memory'].replace(['256GB SSD +  2TB HDD'], '2304')
df['Memory'] = df['Memory'].replace(['128GB SSD +  2TB HDD'], '2176')
df['Memory'] = df['Memory'].replace(['512GB SSD +  2TB HDD'], '2560')
df['Memory'] = df['Memory'].replace(['8GB SSD'], '8')
df['Memory'] = df['Memory'].replace(['508GB Hybrid'], '508')
df['Memory'] = df['Memory'].replace(['64GB Flash Storage +  1TB HDD'], '1088')

# Extract the only Gpu memory space in the column
df['Gpu'] = df['Gpu'].str.extract(r'^(\w+)')

# Drop Unnecessary  columns from the dataset
df=df.drop(['ScreenResolution','laptop_ID','Type_of_screen'],axis=1)
```

```python
# Convert the all the categorical columns into numbers
en = LabelEncoder()
df['Company']= en.fit_transform(df['Company'])
df['Product']= en.fit_transform(df['Product'])
df['TypeName']= en.fit_transform(df['TypeName'])
df['Cpu']= en.fit_transform(df['Cpu'])
df['OpSys']= en.fit_transform(df['OpSys'])
df['Gpu']= en.fit_transform(df['Gpu'])
df['dimen']= en.fit_transform(df['dimen'])
```

```python
df.head()
```

```python
X=df.drop('Price_euros', axis = 1).values
y=df['Price_euros'].values
```

```python
##Taking 30 percent for testing and 70 percent for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 69)
```

```python
##Applying Standard Scaler for normalizing all columns to the same scale
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# **Applying the Bagging**

**Applying on knn**

```python
knn = BaggingRegressor(base_estimator=neighbors.KNeighborsRegressor(n_neighbors = 5) ,n_estimators=10, random_state=0).fit(X_train, y_train)
pred_knn=knn.predict(X_test)
print("Knn Score:", knn.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_knn))
mse_rf = mean_squared_error(np.exp(y_test), np.exp(pred_knn))
print("RMSE using KNeighborsRegressor: {} $ ".format(np.round(np.sqrt(mse_rf)),4))
scores = cross_val_score(knn, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

**Applying on Linear Regression**

```python
regr = BaggingRegressor(base_estimator=linear_model.LinearRegression() ,n_estimators=10, random_state=0).fit(X_train, y_train)
pred_regr = regr.predict(X_test)
print("Linear Regression Score:", regr.score(X_train,y_train))
print("Score of model: ",r2_score(y_test, pred_regr))
mse_regr = mean_squared_error(np.exp(y_test), np.exp(pred_regr))
print("RMSE using Linear Regressor: {} $ ".format(np.round(np.sqrt(mse_regr)),4))
scores = cross_val_score(knn, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

# Applying PCA

```python
pca = PCA(.95)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```python
print(X_train_pca.shape)
print(X_test_pca.shape)
```

```python
def model_evaluator(model , X_train, y_train, X_test, y_test):
    # Train a model
    model.fit(X_train, y_train)
    
    # Make a prediction
    y_pred = model.predict(X_test)
    
    # calculate the model's score
    model_score = model.score(X_train, y_train)
    
    # calculate the r2score
    score_r2 = r2_score(y_test, y_pred)
    
    # calculate the MSE
    mse = np.round(np.sqrt(mean_squared_error(np.exp(y_test), np.exp(y_pred))), 4)
    
    # Cross Validation score
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    
    # Model Accuracy
    model_accuracy = cv_scores.mean()
    
    return [model_score, score_r2, mse, model_accuracy] 
```

# **Without PCA**

```python
# Applying Knn
knn = neighbors.KNeighborsRegressor(n_neighbors = 5)
knn_results = model_evaluator(knn , X_train, y_train, X_test, y_test)

parameters = [{'weights': ['uniform'], 'n_neighbors': [1,2,3,5, 10, 20]}]
knn_gscv = GridSearchCV(knn, param_grid=parameters, cv=5)
knn_gscv.fit(X_train, y_train)
print("knn best parameters:\n", knn_gscv.best_params_)

# Linear Regression
lr = linear_model.LinearRegression()
lr_results = model_evaluator(lr , X_train, y_train, X_test, y_test)

parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
regr_gscv = GridSearchCV(linear_model.LinearRegression(), param_grid=parameters, cv=5)
regr_gscv.fit(X_train, y_train)
print("Linear Regression Best Parameters:\n", regr_gscv.best_params_)

# Ridge
ridge = Ridge()
ridge_results = model_evaluator(ridge , X_train, y_train, X_test, y_test)

alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
parameters = dict(alpha=alpha)
ridge_gscv = GridSearchCV(ridge, param_grid=parameters, cv=5)
ridge_gscv.fit(X_train, y_train)
print("Ridge Best Parameters:\n", ridge_gscv.best_params_)

# Lasso
lasso = Lasso(alpha=0.02)
lasso_results = model_evaluator(lasso , X_train, y_train, X_test, y_test)

lasso_params = {'alpha':[0.02, 0.024, 0.025, 0.026, 0.03]}
lasso_gscv = GridSearchCV(lasso, param_grid=lasso_params, cv=5)
lasso_gscv.fit(X_train, y_train)
print("Lasso Regression best parameters:\n", lasso_gscv.best_params_)

# Polynomial Features
pf = make_pipeline(PolynomialFeatures(2) , linear_model.LinearRegression())
pf_results = model_evaluator(pf , X_train, y_train, X_test, y_test)

param_grid = {'polynomialfeatures__degree': np.arange(5), 'linearregression__fit_intercept': [True, False], 'linearregression__normalize': [True, False]}
polyreg_gscv= GridSearchCV(pf, param_grid, cv=5, scoring='neg_mean_squared_error')
polyreg_gscv.fit(X_train, y_train)
print("Polynomial Regression Best parameters:\n", polyreg_gscv.best_params_)

# Decision Tree
dtr = DecisionTreeRegressor()
dtr_results = model_evaluator(dtr , X_train, y_train, X_test, y_test)

params = {'max_depth': [3,4,5,6], 'min_samples_leaf': [0.02,0.03,0.04,0.06,0.08,0.3,0.4]}
Dec_reg_gscv = GridSearchCV(dtr, param_grid=params, cv=5)
Dec_reg_gscv.fit(X_train, y_train)
print("Decision tree Best Parameters:\n", Dec_reg_gscv.best_params_)

# SVR Linear
svr_linear = SVR(kernel='linear')
svr_linear_results = model_evaluator(svr_linear , X_train, y_train, X_test, y_test)

k=['linear']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
pred_svr_l_gscv = GridSearchCV(svr_linear, param_grid, cv=5)
pred_svr_l_gscv.fit(X_train, y_train)
print("SVR Linear Best Parameters:\n", pred_svr_l_gscv.best_params_)

# SVR Poly
svr_poly = SVR(kernel='poly')
svr_poly_results = model_evaluator(svr_poly , X_train, y_train, X_test, y_test)

k=['poly']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
svr_poly_gscv = GridSearchCV(svr_poly, param_grid, cv=5)
svr_poly_gscv.fit(X_train, y_train)
print("SVR Poly Best Parameters:\n", svr_poly_gscv.best_params_)

# SVR rbf
svr_rbf = SVR(kernel='rbf')
svr_rbf_results = model_evaluator(svr_rbf , X_train, y_train, X_test, y_test)

k=['rbf']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
svr_rbf_gscv = GridSearchCV(svr_rbf, param_grid, cv=5)
svr_rbf_gscv.fit(X_train, y_train)
print("SVR rbf best parameters:\n", svr_rbf_gscv.best_params_)
```

```python
# Applying Knn
knn = neighbors.KNeighborsRegressor(n_neighbors = 5)
pca_knn_results = model_evaluator(knn , X_train_pca, y_train, X_test_pca, y_test)

parameters = [{'weights': ['uniform'], 'n_neighbors': [1,2,3,5, 10, 20]}]
knn_gscv = GridSearchCV(knn, param_grid=parameters, cv=5)
knn_gscv.fit(X_train_pca, y_train)
print("knn best parameters:\n", knn_gscv.best_params_)

# Applying linear Regression 
lr = linear_model.LinearRegression()
pca_lr_results = model_evaluator(lr , X_train_pca, y_train, X_test_pca, y_test)

parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
regr_gscv = GridSearchCV(lr, param_grid=parameters, cv=5)
regr_gscv.fit(X_train_pca, y_train)
print("Linear Regression Best Parameters:\n", regr_gscv.best_params_)

# Applying Ridge
ridge = Ridge()
pca_ridge_results = model_evaluator(ridge , X_train_pca, y_train, X_test_pca, y_test)

alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
parameters = dict(alpha=alpha)
ridge_gscv = GridSearchCV(ridge, param_grid=parameters, cv=5)
ridge_gscv.fit(X_train_pca, y_train)
print("Ridge Best Parameters:\n", ridge_gscv.best_params_)

# Applying the lasso 
lasso = Lasso(alpha=0.02)
pca_lasso_results = model_evaluator(lasso , X_train_pca, y_train, X_test_pca, y_test)

lasso_params = {'alpha':[0.02, 0.024, 0.025, 0.026, 0.03]}
lasso_gscv = GridSearchCV(lasso, param_grid=lasso_params, cv=5)
lasso_gscv.fit(X_train_pca, y_train)
print("Lasso Regression best parameters:\n", lasso_gscv.best_params_)

# Applying the Polynomial Features
pf = make_pipeline(PolynomialFeatures(2) , linear_model.LinearRegression())
pca_pf_results = model_evaluator(pf , X_train_pca, y_train, X_test_pca, y_test)

param_grid = {'polynomialfeatures__degree': np.arange(5), 'linearregression__fit_intercept': [True, False], 'linearregression__normalize': [True, False]}
polyreg_gscv= GridSearchCV(pf, param_grid, cv=5, scoring='neg_mean_squared_error')
polyreg_gscv.fit(X_train_pca, y_train)
print("Polynomial Regression Best parameters:\n", polyreg_gscv.best_params_)

# Applying the Decision tree Classifier
dtr = DecisionTreeRegressor()
pca_dtr_results = model_evaluator(dtr , X_train_pca, y_train, X_test_pca, y_test)

params = {'max_depth': [3,4,5,6], 'min_samples_leaf': [0.02,0.03,0.04,0.06,0.08,0.3,0.4]}
Dec_reg_gscv = GridSearchCV(dtr, param_grid=params, cv=5)
Dec_reg_gscv.fit(X_train_pca, y_train)
print("Decision tree Best Parameters:\n", Dec_reg_gscv.best_params_)

# SVR linear
svr_linear = SVR(kernel='linear')
pca_svr_linear_results = model_evaluator(svr_linear , X_train_pca, y_train, X_test_pca, y_test)

k=['linear']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
pred_svr_l_gscv = GridSearchCV(svr_linear, param_grid, cv=5)
pred_svr_l_gscv.fit(X_train_pca, y_train)
print("SVR Linear Best Parameters:\n", pred_svr_l_gscv.best_params_)

# SVR Polynomial
svr_poly = SVR(kernel='poly')
pca_svr_poly_results = model_evaluator(svr_poly , X_train_pca, y_train, X_test_pca, y_test)

k=['poly']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
svr_poly_gscv = GridSearchCV(svr_poly, param_grid, cv=5)
svr_poly_gscv.fit(X_train_pca, y_train)
print("SVR Poly Best Parameters:\n", svr_poly_gscv.best_params_)

# SVR rbf
svr_rbf = SVR(kernel='rbf')
pca_svr_rbf_results = model_evaluator(svr_rbf , X_train_pca, y_train, X_test_pca, y_test)

k=['rbf']
c= range(1,10)
param_grid=dict(kernel=k, C=c)
svr_rbf_gscv = GridSearchCV(svr_rbf, param_grid, cv=5)
svr_rbf_gscv.fit(X_train_pca, y_train)
print("SVR rbf best parameters:\n", svr_rbf_gscv.best_params_)
```
